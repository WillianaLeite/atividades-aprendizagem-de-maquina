# Iris

```python
data = load_iris()
df_iris = pd.DataFrame(data.data, columns=data.feature_names)
df_iris['target'] = pd.Series(data.target)
```

```python
realizations_iris_q = p.pipeline(
     df_iris, 
     model_name='classifier_bayes_discriminant',
     discrimant='quadratico',
     col_target='target',
     train_size=0.8,
     n_realizations=20
)
```

```python
realizations_iris_l_a = p.pipeline(
     df_iris, 
     model_name='classifier_bayes_discriminant',
     discrimant='linear', 
     type_cov='all_data',
     col_target='target',
     train_size=0.8,
     n_realizations=20
)
```

```python
realizations_iris_l_p = p.pipeline(
     df_iris, 
     model_name='classifier_bayes_discriminant',
     discrimant='linear', 
     type_cov='pool',
     col_target='target',
     train_size=0.8,
     n_realizations=20
)
```

```python
best_realization_q = p.select_best_realization(realizations_iris_q)
best_realization_l_p = p.select_best_realization(realizations_iris_l_p)
```

### only 2 attributes to plot the decision surface

```python
sns.pairplot(df_iris, hue='target', diag_kind="hist", palette={0: '#0079fa', 1: '#fa6c00', 2: '#14a620'},markers=["o", "s", "D"])
```

```python
# Retreinando com melhor conjunto, porque agora s utilizaremos duas colunas
bayes = ClassifierBayesDiscriminant(discrimant='quadratico')
bayes.fit(
    best_realization_q['train'][['sepal length (cm)', 'petal length (cm)']].copy(),
    best_realization_q['train']['target']
)

plots.plot_decision_boundary(
    
    best_realization_q['test'][['sepal length (cm)', 'petal length (cm)', 'target']].copy(),
    list_features = ['sepal length (cm)', 'petal length (cm)'],
    col_target='target',
    model = bayes,
    task = 'multiclass_classification',
    h=.1
)
```

```python
bayes = ClassifierBayesDiscriminant(discrimant='linear', type_cov='pool')
bayes.fit(
    best_realization_l_p['train'][['sepal length (cm)', 'petal length (cm)']].copy(),
    best_realization_l_p['train']['target']
)

plots.plot_decision_boundary(
    
    best_realization_l_p['test'][['sepal length (cm)', 'petal length (cm)', 'target']].copy(),
    list_features = ['sepal length (cm)', 'petal length (cm)'],
    col_target='target',
    model = bayes,
    task = 'multiclass_classification',
    h=.1
)
```

```python
list_col_names = [
      'erythema',
      'scaling',
      'definite borders',
      'itching',
      'koebner phenomenon',
      'polygonal papules',
      'follicular papules',
      'oral mucosal involvement',
      'knee and elbow involvement',
     'scalp involvement',
     'family history',
     'melanin incontinence',
     'eosinophils in the infiltrate',
     'PNL infiltrate',
     'fibrosis of the papillary dermis',
     'exocytosis',
     'acanthosis',
     'hyperkeratosis',
     'parakeratosis',
     'clubbing of the rete ridges',
     'elongation of the rete ridges',
     'thinning of the suprapapillary epidermis',
     'spongiform pustule',
     'munro microabcess',
     'focal hypergranulosis',
     'disappearance of the granular layer',
     'vacuolisation and damage of basal layer',
     'spongiosis',
     'saw-tooth appearance of retes',
     'follicular horn plug',
     'perifollicular parakeratosis',
     'inflammatory monoluclear inflitrate',
     'band-like infiltrate',
     'age', 
     'target']

df_dermatology = pd.read_csv('./datasets/dermatology/dermatology.data', names = list_col_names)
df_dermatology['age'] =  pd.to_numeric(df_dermatology['age'], downcast='integer', errors='coerce')

df_dermatology['target'] = df_dermatology['target'].apply(lambda y: (y - 1)) # Formatando a varivel resposta para comear do 0

df_dermatology = df_dermatology.fillna(df_dermatology.mean()).astype('int32') # Preenchendo valores null com mdia
```
